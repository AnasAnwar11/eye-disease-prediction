```python
import os
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
read_directory = "Glaucoma\Train\Glaucoma_Positive"
write_directory = "Glaucoma_CLAHE\Glaucoma_Positive"

if os.path.isdir(write_directory) is False:
    os.makedirs(write_directory)

for file in os.listdir(read_directory):
    f_read = os.path.join(read_directory, file)
    
#     print(f_read)
    
    img = cv2.imread(f_read, 1)
    
    img = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)

    l, a, b = cv2.split(img)

    # create a CLAHE object (Arguments are optional).
    clahe = cv2.createCLAHE(clipLimit=5.5, tileGridSize=(20, 20))
    eq_l = clahe.apply(l)

    merged = cv2.merge([eq_l, a, b])

    merged_eq_img = cv2.cvtColor(merged, cv2.COLOR_LAB2BGR)
    
    merged_eq_img = cv2.resize(merged_eq_img, (224, 224))
    
    f_write = os.path.join(write_directory, file)

    cv2.imwrite(f_write, merged_eq_img)
    
#     cv2.imshow('Image', img)
#     cv2.waitKey(0)
#     cv2.destroyAllWindows()
```
